```
import cv2
import numpy as np
from os import listdir
import string
from sklearn.preprocessing import LabelEncoder
from keras import Sequential
from keras.layers import *
from keras.utils import to_categorical
from PIL import Image
from sklearn.utils import shuffle
from keras.optimizers import RMSprop
import tensorflow
```

```

mypath_labels = '/content/gdrive/My Drive/OID/Dataset/train/Weapon_Human body/Label/'
mypath_img = '/content/gdrive/My Drive/OID/Dataset/train/Weapon_Human body/'
# h = 244
# w = 244
# print(244 / 32)
# print(15//5)
# print(w//{0:32, 1:16, 2:8}[0])
# math = h//{0:32, 1:16, 2:8}[0]
# print(h//{0:32, 1:16, 2:8}[0])
def load_images(wide, height, path):

    array = np.empty([0, wide, height, 3])

    names = [p for p in listdir(path)if p[-4:] == ".jpg"]

    for i in names:
        img = cv2.imread(path + i) / 255.
        img = cv2.resize(img, (wide, height), interpolation = cv2.INTER_AREA)
        img = np.expand_dims(img, 0)
        array = np.concatenate((array, img), axis=0)
        print(i)
    return array
def load_labels(path):
    col_num = 0
    delimiter = " " 
    array_name = np.empty(0)
    names = [p for p in listdir(path)if p[-4:] == ".txt"]
    for name in names:
        file = open(path + name)
        data = file.readline().split(delimiter)[col_num]
        print(data) 
        array_name = np.append(array_name, data.strip('\n')) 
        #print(path + name)
    return array_name

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D

modle = Sequential()
modle.add(Conv2D(32, kernel_size=(3, 3), input_shape=(128, 128, 3), activation="relu", padding="same"))
modle.add(Conv2D(32, kernel_size=(3, 3), activation="relu", padding="same"))
modle.add(MaxPooling2D(pool_size=(2, 2)))
modle.add(Dropout(0.25))

modle.add(Conv2D(64, kernel_size=(3, 3), activation="relu", padding="same"))
modle.add(Conv2D(64, kernel_size=(3, 3), activation="relu", padding="same"))
modle.add(MaxPooling2D(pool_size=(2, 2)))
modle.add(Dropout(0.25))

modle.add(Conv2D(128, kernel_size=(3, 3), activation="relu", padding="same"))
modle.add(Conv2D(128, kernel_size=(3, 3), activation="relu", padding="same"))
modle.add(MaxPooling2D(pool_size=(2, 2)))
modle.add(Dropout(0.25))

modle.add(Flatten())
modle.add(Dense(256, activation="relu"))
modle.add(Dense(128, activation="relu"))
modle.add(Dense(1, activation="sigmoid"))
rmsprop = RMSprop(learning_rate=0.0001, decay=1e-6, momentum=0.5, centered=True)

modle.compile(optimizer=rmsprop , loss="binary_crossentropy", metrics=["accuracy"])

import numpy as np
X_train = np.load('/content/gdrive/My Drive/models_builds/X_train.npy')
Y_train = np.load('/content/gdrive/My Drive/models_builds/Y_train.npy')
Y_train_normal = Y_train
label_encoder = LabelEncoder()
Y_train = label_encoder.fit_transform(Y_train)
Y_train = Y_train == 3
print(Y_train.shape)
print(Y_train_normal)
print(Y_train)
print(X_train.shape)
print(tensorflow.__version__)
```
